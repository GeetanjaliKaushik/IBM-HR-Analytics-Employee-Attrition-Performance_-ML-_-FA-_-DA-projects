```python
import pandas as pd
import numpy as np
import seaborn as sns

from scipy.stats import chi2_contingency, ttest_ind, mannwhitneyu
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

sns.set(style="whitegrid")
```

```python
# Make sure the attrition column is consistent: 'Attrition' contains 'Yes'/'No'
IBM_HR_data['AttritionFlag'] = IBM_HR_data['Attrition'].map({'Yes':1, 'No':0})
if IBM_HR_data['AttritionFlag'].isnull().any():
    raise ValueError("Unexpected values in Attrition column. Check unique values: " + str(IBM_HR_data['Attrition'].unique()))

# Age groups: adjust bins if you prefer different ranges
IBM_HR_data['AgeGroup'] = pd.cut(IBM_HR_data['Age'], bins=[17,30,40,50,60,100], labels=['18-30','31-40','41-50','51-60','61+'])

# Optional: create a tenure column if not present (some IBM datasets have YearsAtCompany)
if 'YearsAtCompany' in IBM_HR_data.columns:
    IBM_HR_data['TenureYears'] = IBM_HR_data['YearsAtCompany']
else:
    # If you have HireDate use pd.to_datetime and today's date to compute tenure
    pass

display(IBM_HR_data[['Attrition','AttritionFlag','Age','AgeGroup']].head())
```

```python
# === Cell 4: Overall turnover & demographic breakdowns ===
overall_turnover = IBM_HR_data['AttritionFlag'].mean()
print(f"Overall turnover (attrition) rate: {overall_turnover:.2%}")

# By Gender
by_gender = IBM_HR_data.groupby('Gender')['AttritionFlag'].agg(['count','sum','mean']).reset_index().rename(columns={'sum':'Attrited','mean':'AttritionRate'})
by_gender['AttritionRate_pct'] = (by_gender['AttritionRate']*100).round(2)
display(by_gender.sort_values('AttritionRate', ascending=False))

# By AgeGroup
by_age = IBM_HR_data.groupby('AgeGroup')['AttritionFlag'].agg(['count','sum','mean']).reset_index()
by_age['AttritionRate_pct'] = (by_age['mean']*100).round(2)
display(by_age.sort_values('mean', ascending=False))

# By Education (if present)
if 'Education' in IBM_HR_data.columns:
    by_edu = IBM_HR_data.groupby('Education')['AttritionFlag'].agg(['count','sum','mean']).reset_index()
    display(by_edu.sort_values('mean', ascending=False))

# By Department
by_dept = IBM_HR_data.groupby('Department')['AttritionFlag'].agg(['count','sum','mean']).reset_index()
by_dept['AttritionRate_pct'] = (by_dept['mean']*100).round(2)
display(by_dept.sort_values('mean', ascending=False))

# By JobRole
by_role = IBM_HR_data.groupby('JobRole')['AttritionFlag'].agg(['count','sum','mean']).reset_index()
by_role['AttritionRate_pct'] = (by_role['mean']*100).round(2)
display(by_role.sort_values('mean', ascending=False))

```

```python
# visualizations ===
plt.figure(figsize=(8,4))
sns.barplot(data=by_dept.sort_values('mean', ascending=False), x='Department', y='mean')
plt.title('Attrition Rate by Department')
plt.ylabel('Attrition Rate')
plt.xlabel('')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,4))
sns.barplot(data=by_age.sort_values('mean', ascending=False), x='AgeGroup', y='mean', order=['18-30','31-40','41-50','51-60','61+'])
plt.title('Attrition Rate by Age Group')
plt.ylabel('Attrition Rate')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=IBM_HR_data)
plt.title('Monthly Income distribution by Attrition')
plt.show()
```
